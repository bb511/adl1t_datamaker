```python
from pathlib import Path

import awkward as ak

from adl1t_datamaker.convert import loader
from adl1t_datamaker.convert import plots
from adl1t_datamaker.convert.terminal_colors import tcols
```

```python
select_feats = {
  'muons': [
    'muonIPhiAtVtx', 'muonIEt', 'muonQual',
    'muonIEtaAtVtx', 'muonChg', 'muonTfMuonIdx',
    'muonIPhi', 'muonIEta', 'muonIEtUnconstrained'
  ],
  'jets': ['jetIEt', 'jetIEta', 'jetIPhi', 'jetHwQual', 'jetRawEt'],
  'egammas': ['egIEt', 'egIEta', 'egIPhi', 'egIso'],
  'taus': ['tauIEt', 'tauIEta', 'tauIPhi', 'tauIso'],
  'ET': ['Et', 'ETTEM'],
  'HT': ['Et', 'tower_count'],
  'MET': ['Et', 'phi'],
  'MHT': ['Et', 'phi'],
  'FET': ['Et', 'phi'],
  'FHT': ['Et', 'phi'],
  'cicada': ['CICADAScore'],
  'event_info': 'none',
  'seeds': 'none'
}

data_path = Path('/data/deodagiu/adl1t_data/parquet_files/EphZB_2025B_run392642')
# data_path = Path('/data/deodagiu/adl1t_data/parquet_files/EphZB_2025G_run398183')
# data_path = Path('/data/deodagiu/adl1t_data/parquet_files/EphZB_2024E_run381148-381149')
data = loader.Parquet2Awkward(data_path, select_feats=select_feats)
data_name = data_path.stem
plots_folder = Path('/data/deodagiu/adl1t_datamaker/notebooks/zero_energy_plots') / data_name
```

```python
muons_data = data['muons']
context = {feature: muons_data[feature] for feature in muons_data.fields}
context.keys()

# Select objects with 0 pT.
zero_energy_muons = muons_data[muons_data.muonIEt == 0]
obj_folder = plots_folder / 'muons'
obj_folder.mkdir(exist_ok=True, parents=True)
for feature in zero_energy_muons.fields:
    plots.plot_hist(zero_energy_muons[feature], feature, obj_folder)
```

```python
HT_data = data['HT']
context = {feature: HT_data[feature] for feature in HT_data.fields}
print(context.keys())

# Select objects with 0 pT.
zero_energy_HT = HT_data[HT_data.Et == 0]
count = ak.sum(
    ak.any(
        [ak.fill_none(ak.num(zero_energy_HT[f]) > 0, False) for f in ak.fields(zero_energy_HT)],
        axis=0
    )
)
print(count)
obj_folder = plots_folder / 'HT'
obj_folder.mkdir(exist_ok=True, parents=True)
for feature in zero_energy_HT.fields:
    plots.plot_hist(zero_energy_HT[feature], feature, obj_folder)
```
